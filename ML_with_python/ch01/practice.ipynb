```python
import scipy as sp
```

```python
data = sp.genfromtxt("/Users/jhyun/Desktop/Programming_Bundle/Modu/BuildingMachineLearningSystemsWithPython-master/ch01/data/web_traffic.tsv", delimiter="\t")
```

```python
print(data[:10])
```

```python
x = data[:, 0]
y = data[:, 1]
```

```python
sp.sum(sp.isnan(y))
```

```python
x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(x, y, s = 10)
plt.title("Web trffic over the last month")
plt.xlabel("time")
plt.ylabel("hits/hour")
plt.xticks([w*7*24 for w in range(10)], ['week %i' % w for w in range(10)])
plt.autoscale(tight=True)
plt.grid(True, linestyle='-', color='0.75')
plt.show()
```

```python
def error(f, x, y):
    return sp.sum((f(x)-y)**2)
```

```python
fp1, residuals, rank, sv, rcond = sp.polyfit(x, y, 1, full=True)
```

```python
print("Models parameters: %s" %fp1)
```

```python
print(residuals)
```

```python
f1 = sp.poly1d(fp1)
print(error(f1, x, y))
```

```python
fx = sp.linspace(0, x[-1], 1000)
plt.plot(fx, f1(fx), linewidth = 4)
plt.legend(["d=%i" %f1.order], loc="upper left")
plt.scatter(x, y, s = 10)
plt.title("Web trffic over the last month")
plt.xlabel("time")
plt.ylabel("hits/hour")
plt.xticks([w*7*24 for w in range(7)], ['week %i' % w for w in range(7)])
plt.grid(True, linestyle='-', color='0.75')
plt.show()
```
